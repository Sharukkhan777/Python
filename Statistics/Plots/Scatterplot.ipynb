### Import required libraries

### Get Started

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
import numpy as np
```

```python
N = 1000
example_x = np.random.rand(N)
example_y = np.random.rand(N)
```

```python
# ----------trace means name of the collection-----------
trace = go.Scatter(
    x = example_x,
    y = example_y,
    mode = 'markers')  # ---help(go.Scatter.mode)--- to see the details
```

```python
data = [trace]
iplot(data)
```
