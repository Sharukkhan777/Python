## 1) Simple get and post

```python
from flask import Flask, render_template, request   #>>>>>> here we add request, 
                                                            # whether to select GET or POST
app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html")   

@app.route("/login", methods = ["GET","POST"])
def login():
    if request.method == "POST":
        user = request.form["nm"]
        return redirect(url_for("user", usr = user)) # see redirect file for more info
    else:
        return render_template("login.html")

@app.route("/<usr>")
def user(usr):
    return f"<h1>{usr}</h1>"

if __name__ =="__main__":
    app.run()
```
