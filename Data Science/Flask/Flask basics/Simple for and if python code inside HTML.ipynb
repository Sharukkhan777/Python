## 1) For loop and if statement

```python
from flask import Flask, render_template   #>>>>>> here we add render_template

app = Flask(__name__)

@app.route("/<name>")
def home(name):
    return render_template("index.html")

if __name__ =="__main__":
    app.run()
```

#### Include the HTML file 

Note:
    * Save this below html file as ---index.html---(Recommended for home page)
    * Save in the folder name ---"templates"---, which we have to create in where our python file located

```python
<!doctype html>
<html>
<head>
    <title> Home Page </title>
</head>
<body>
    {% for x in range(10) %}   <!--Note: no ":"-->
        {% if x%2 == 0}
            <p>"test"</p>
        {% endif %}
    {% endfor %}
</body>
</html>
```
